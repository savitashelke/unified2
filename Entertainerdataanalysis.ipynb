# How many Entertainers had a Breakthrough in the year 2000 or later?

```python
breakthrough_years =final_df['Year of Breakthrough/#1 Hit/Award Nomination'].value_counts()
```

```python
breakthrough_years 
```

```python
entertainer_names = final_df['Entertainer']
```

```python
breakthrough_years = final_df['Year of Breakthrough/#1 Hit/Award Nomination']
```

```python
indices = np.where(breakthrough_years >= 2000)[0]
```

```python
for i in range(4):
    index = indices[i]
    name = entertainer_names[index]
    year = breakthrough_years[index]
    print(f"Entertainer: {name}, Breakthrough Year: {year}")
```

# Who is the oldest Entertainer in the dataset ?

```python
birth_years = final_df['Birth Year']
```

```python
oldest_index = np.argmin(birth_years)
```

```python
oldest_entertainer = final_df.loc[oldest_index, 'Entertainer']
```

```python
print(" The Oldest Entertainer is:", oldest_entertainer)
```

# Who is youngest Entertainer in dataset?

```python
youngest_index = np.argmax(birth_years)
```

```python
youngest_entertainer = final_df.loc[youngest_index, 'Entertainer']
```

```python
print(" The youngest Entertainer is:", youngest_entertainer)
```

# How many Entertainers have won an Oscar, Grammy or Emmy ??

```python
entertainer_names = final_df['Entertainer']
```

```python
Oscar_won = final_df['Year of First Oscar/Grammy/Emmy']
```

```python
indices = np.where(Oscar_won > 0)[0]
```

```python
print("Entertainers who have won an Oscar:")
for index in indices:
    name = entertainer_names[index]
    print(name)
```

```python
num_oscar_winners = np.sum(Oscar_won > 0)
```

```python
num_oscar_winners
```

```python
print("Number of entertainers who have won an Oscar, Grammy or Emmy :", num_oscar_winners)
```
